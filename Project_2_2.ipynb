```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

```python
df =pd.read_csv('employee_data.csv')
```

```python
x =np.linspace(0,1,100)
plt.plot(x,x>0.5,color ='k')
```

```python
np.random.seed(555)
noise =np.random.uniform(-0.2,0.2,100)
#target_variable
y =((x+noise)>0.5).astype(int)
```

```python
plt.scatter(x,y)
plt.plot(x,x>0.5,'k--')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
```

```python
X =x.reshape(100,1)
print(X[:5])
```

```python
y.shape
```

```python
model =LinearRegression()
model.fit(X,y)
plt.scatter(X,y)
plt.plot(X,model.predict(X),'k--')
plt.show()
```

```python
model =LogisticRegression()
model.fit(X,y)
pred =model.predict_proba(X[:10])
```

```python
pred
```

```python
#logisticRegression
pred = model.predict_proba(X)
pred =[p[1] for p in pred]
plt.scatter(X,y)
plt.plot(X,pred,'k--')
plt.show()
```

```python
def fit_and_plot_classifier(clf):
    clf.fit(X,y)
    
    pred = clf.predict_proba(X)
    pred = [p[1] for p in pred]
    
    plt.scatter(X,y)
    plt.plot(X,pred,'k--')
    plt.show()
    return clf ,pred
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression())
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression(C=0.25))
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression(C=0.01))
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression(C=4))
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression(C=10000))
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression(penalty ='l1'))
```

```python
clf,pred = fit_and_plot_classifier(LogisticRegression(penalty ='l1',C=4))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RandomForestClassifier()
```

```python
clf,pred = fit_and_plot_classifier(RandomForestClassifier(random_state =123))
```

```python
for n_trees in [20,50,100,200]:
    print('Number of trees :',n_trees)
    fit_and_plot_classifier(RandomForestClassifier(random_state =123,n_estimators =n_trees))
```

```python
for leaf_size in [2,5,10,20]:
    print('Minimum Leaf Size:',leaf_size)
    fit_and_plot_classifier(RandomForestClassifier(random_state =123,min_samples_leaf =leaf_size))
    
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GradientBoostingClassifier()
```

```python
clf,pred = fit_and_plot_classifier(GradientBoostingClassifier(random_state =123))
```

```python
for n_trees in [10,20,50,200]:
    print('Number of trees :',n_trees)
    fit_and_plot_classifier(GradientBoostingClassifier(random_state =123,n_estimators =n_trees))
```

```python
for depth in [1,2,5,10]:
    print('Max Depth :',depth)
    fit_and_plot_classifier(GradientBoostingClassifier(random_state =123,max_depth=depth))
```
